```python
#feature mapping
def map_feature(X1,X2):
    X1=np.atleast_1d(X1)
    X2=np.atleast_1d(X2)
    degree=6
    out=[]
    for i in range(1,degree+1):
      for j in range(i+1):
        out.append((X1**(i-j) * (X2**j)))
    return np.stack(out, axis=1)
```

```python
#computing cost
def compute_cost(X,y,w,b,lambda_=1):
    total_cost = 0
    size, features = X.shape
    for i in range(size):
        f = np.dot(X[i], w) + b
        y_predict = sigmoid(f)
        total_cost += -(y[i])*np.log(y_predict) - (1 - y[i])*np.log(1 - y_predict)

    # for j in range(features):
    #     total_cost += (lambda_/2)*(w[j]**2)

    total_cost /= size
    return total_cost
```

```python
#compute regularised cost
def compute_cost_reg(X, y, w, b, lambda_ = 1):

    size, features = X.shape
    cost = 0
    total_cost = compute_cost(X, y, w, b)

    for j in range(features):
        cost += w[j]**2

    cost *= (lambda_)/(2*size)
    total_cost += cost
    return total_cost
```

```python
#compute gradient
def compute_gradient(X, y, w, b, lambda_=None):
    size, features = X.shape
    dj_db = 0
    dj_dw = np.zeros((features, ))
    f = sigmoid(np.dot(X, w) + b)

    for i in range(size):
        delta = f[i] - y[i]
        dj_db += delta
        for j in range(features):
            dj_dw[j] += delta*X[i, j]
    dj_db /= size
    dj_dw /= size
    return dj_db,dj_dw
```

```python
#compute regularised gradient
def compute_gradient_reg(X, y, w, b, lambda_=1):
    size, features = X.shape
    dj_db, dj_dw = compute_gradient(X, y, w, b)
    correction = np.zeros((features, ))

    for j in range(features):
        correction[j] += w[j]

    correction *= lambda_/size
    dj_dw += correction

    return dj_db,dj_dw
```
